```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
```

## 1.Data Collection

```python
data = pd.read_csv('Datasets/LabTAT.csv')
data
```

## 2.Data Analysis

```python

#this is ANOVA test case as it involves 4 independent samples means
#lets check the assumptions of one way ANNOVA to proceed further

#normal distribution
for i in data.columns:
    sns.distplot(data[i],hist=False)
    plt.show()
```

**this distribution looks pretty normal***

```python
#equal variance check

print(data['Laboratory 1'].var(),data['Laboratory 2'].var(),data['Laboratory 3'].var(),data['Laboratory 4'].var())

plt.boxplot(data)
plt.show()
```

## 2.Hypothesis Formulation

**H0: xÌ…1 = xÌ…2 = xÌ…3 = xÌ…4** 

**Ha: At least one of the group mean differs**

**significance level Î± = 0.05**

## 3. P value caculation

```python
Î± = 0.05
f_oneway = stats.f_oneway(data.iloc[:,0],data.iloc[:,1],data.iloc[:,2],data.iloc[:,3])

if f_oneway.pvalue < Î±:
    print('We Reject Null hypothesis')
    print('Atleast one of the groups have different mean ')
else:
    print('We fail to reject NUll hypothesis')
    print('there is no significant difference among group means')
```

## using Cosine similaity to get User-User collebrative filtering

```python
#creating a pivot table
pivot = pd.pivot_table(data = data,values='Book.Rating',index='User.ID',columns='Book.Title').fillna(0)
pivot.head()
```

```python
cosine_df = pd.DataFrame(1 - pairwise_distances(X = pivot,metric='cosine')).round(2)
cosine_df.columns = data['User.ID'].unique()
cosine_df.index   = data['User.ID'].unique()
cosine_df
```

```python
np.fill_diagonal(a = cosine_df.to_numpy(),val = 0)
```

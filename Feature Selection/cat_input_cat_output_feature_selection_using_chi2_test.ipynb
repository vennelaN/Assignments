## Catagorical input and Catgorical output
### here we will use Chi-square test from sklear which Will calculate Chi-square between each feature and the target and select the desired number of features with the best Chi-square scores

**following conditions have to be met: the variables have to be categorical, sampled independently and values should have an expected frequency greater than 5**

```python
import pandas as pd
pd.set_option('display.max_columns',None)
from sklearn.feature_selection import chi2,SelectKBest
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../../Exlr_Assignments/Datasets/Zoo.csv')
data.head()
```

```python
del data['animal name']
```

```python
data = pd.get_dummies(data=data,columns=['legs'],drop_first=True)
data.head()
```

```python
data.shape
```

```python
X = data.drop(labels='type',axis=1)
y = data['type']
```

```python


chi2_features = SelectKBest(chi2,k=15).fit(X,y)
print(X.columns)
print(X.shape)
print(X.columns[chi2_features.get_support()])
X_selected = chi2_features.transform(X)
print(X_selected.shape)


```
